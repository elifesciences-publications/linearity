```python
def linearModel(x, a=1):
    # Linear model
    return (a*x)

def DN_model(x,a=1):
    # Divisive normalization model
    #return x - a*(x**2)/(b+x)
    return x/(a*x+1)

def DN_1_model(x,a=1):
    # Divisive normalization model
    #return x - a*(x**2)/(b+x)
    return a*x/(x+1)

def DN_2_model(x,a=1,b=1):
    # Divisive normalization model
    #return x - a*(x**2)/(b+x)
    return x/(a*x+b)
```
