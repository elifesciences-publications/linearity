```python
from sympy import *
```

```python
init_session()
```

```python
t, P, e_r, e_d, delta_e, rho_e, g_e, i_r, i_d, delta_i, rho_i, g_i, b = symbols('t P \\tau_{er} \\tau_{ed} \\delta_e \\rho_e \\bar{g}_e \\tau_{ir} \\tau_{id} \\delta_i \\rho_i \\bar{g}_i \\beta', positive=True, real=True)
```

```python
C = g_e*(P*(tau_i/(tau_i-1))*exp((t-delta_i)/i_r)*(1 - exp(-(t-delta_i)/(i_r*(1+(1/tau_i))))) - (tau_e/(tau_e-1))*exp(t/e_r)*(1 - exp(-t/(e_r*(1+(1/tau_e))))))
```

```python
C
```

```python
C_prime = diff(C,t)
```

```python
C_prime
```

```python
C_check = C_prime.subs({P:2, tau_i:10, tau_e:4, delta_i:1/g_e, i_r : 2.8, e_r: 5, t: 4})
```

```python
simplify(C_check)
```
